# Inventory Management System
# Adding Products

```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()
items = json.loads(r)
```

```python
items
```

# Add items

```python
add_prod = input("Would you like to add new items?")
if add_prod.lower() == 'y':
    number_of_items = int(input("How many products would you like to add? "))
    while (number_of_items):
        prod_id = input("Enter product id:")
        name = input("Enter name:")
        price = float(input("Enter price:"))
        quantity = int(input("Enter quantity:"))
        category = input("Enter category:")
        discount = int(input("Enter discount:"))
        number_of_items -= 1
        items[prod_id] = {'Name': name, 'Price': price, 'Quantity': quantity, 'Category': category, 'Discount': discount}
        js = json.dumps(items)
        fd = open("record.json",'w')
        fd.write(js)
        fd.close()
        print("Items added successfully")
else:
    print("No items added")
```

# Delete items

```python
delete_prod = input("Would you like to remove any item?")
if delete_prod.lower() == 'y':
    remove_items = int(input("How many products would you like to remove? "))
    while (remove_items):
        prod_id = input("Enter product id:")
        remove_items -= 1
        del items[prod_id]
        print("Items removed successfully")
else:
    print("No items removed")
```

# Buy items

```python
sales = {}
customer_name=input("Enter your name: ")
sales[customer_name]={"items":[],"total":0}
while True:
    prod_id = str(input("Enter the product_Id: "))
    quantity = int(input("Enter the quantity: "))
    if (items[prod_id]['Quantity'] > 0):
        print("Product: ", items[prod_id]['Name'])
        print("Price: ", items[prod_id]['Price'])
        print("Total amount: ", items[prod_id]['Price'] * quantity)
        items[prod_id]['Quantity'] = items[prod_id]['Quantity'] - quantity
        sales[customer_name]['items'].append([
                    {
                     "itemId": prod_id,"Quantity": quantity
                },
        ])
        sales[customer_name]['total'] += (items[prod_id]['Price'] * quantity)
    else:
        print("Sorry the item is no more available")
        items[prod_id]['Quantity'] = 0
    if input("Do you want to buy more Y/N: ").lower()=='n':
        print("Thankyou {} for shopping!" .format(customer_name))
        break
```
